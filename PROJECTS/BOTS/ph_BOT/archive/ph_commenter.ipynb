```python
def pull_video_ids_from_url(): #{
    # [2019-20-22]\\url = input("Enter a website to extract the ID's from:")
    url = "http://www.pornhub.com"
    # [2019-20-22]\\r = requests.get("http://" + url)
    r = requests.get(url)
    data = r.text
    soup = BeautifulSoup(data)
    # EMPTY LIST TO HOLD DATA
    id_list = []
    for li in soup.find_all('li'): #{
        if str(li.get('id')) != "None": #{
            # PRINT TO OUTPUT
            logging.info(li.get('id'))
            # append to LIST
            id_list.append(str(li.get('id')))
        #}
        """
        else: #{
            print("...")
        #}
        """
    #}
    logging.info("LEN OF LIST == " + str(len(id_list)))
    # CREATE DATAFRAME FROM LIST
    ids_df = pd.DataFrame(data=id_list, columns=['id'], dtype=np.str)
    # RETURN DATAFRAME
    return ids_df
#}
```

```python
def pull_links_from_ph(the_url): #{
    # RE-INSTANTIATE GLOBALS
    global link_list
    # [2019-10-22]\\url = input("Enter a website to extract the URL's from: ")
    URL = str(the_url) #"http://www.pornhub.com"
    # [2019-10-22]\\r = requests.get("http://" + url)
    r = requests.get(URL)
    data = r.text
    soup = BeautifulSoup(data)
    for link in soup.find_all('a'): #{
        # WRTIE TO LOG
        logging.info(link.get('href'))
        # CREATE VARIABLE OF POSSIBLE LINK
        soup_link = str(link.get('href'))
        # CHECK TO SEE IF VALID LINK
        if fnmatch.fnmatch(soup_link, "*view_video.php?*"): #{
            # IS A LEGIT LINK... now check if fully complete or not
            #print(soup_link)
            if fnmatch.fnmatch(soup_link, "*https://www.pornhub.com/*"): #{
                # IT IS A COMPLETE LINK! append to list?
                logging.info(soup_link)
                link_list.append(soup_link)
            #}
            else: #{
                # ITS NOT COMPLETE! SO COMPLETE IT!
                corrected_link = str("https://www.pornhub.com" + soup_link)
                logging.info("CORRECTED LINK == " + corrected_link)
                link_list.append(corrected_link)
            #}
        #}
        else: #{
            logging.info("NOT LEGIT LINK!")
        #}
        
    #}
#}
```
