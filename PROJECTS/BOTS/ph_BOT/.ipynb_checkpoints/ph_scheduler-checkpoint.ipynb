## *create DICT*

```python
class EventScheduler(): #{
    
    def __init__(self, cmd_file): #{
        self.cmd_file = cmd_file
        # [2019-11-07]//self.time_check = pd.Timedelta(value="12 hours 15 minutes")
        self.single_video_time_check = pd.Timestamp(ts_input="16:30:00")
        print(str(self.single_video_time_check))
        self.begin_loop()
    #}
    
    def begin_loop(self): #{
        #####################################
        # BEGIN INFINITE LOOP
        while 1: #{
            # PULL TIMESTAMP
            current_timestamp = self.pull_current_time()
            # TIMESTAMP STRING
            ts_str = str(self.pull_current_time())
            print("CURRENT TIME == " + ts_str)
            ############################
            # PULL / CREATE TIME CHECKS:
            print("TIME_CHECK == " + str(self.single_video_time_check))
            ############################
            # COMPARE / TIME-CHECK:
            if current_timestamp > self.single_video_time_check: #{
                print("\t\t\t >>>>> TIME HAS PASSED!")
            #}
            else: #{
                print("\t\t\t >>>>> HAS NOT PASSED!\n\t" 
                      + str(self.single_video_time_check - current_timestamp) 
                      + " REMAINING !")
            #}
            self.countdown(60)
        #}
        #####################################
    #}
    
    def countdown(self, n): #{
        while n > 0: #{
            print(n)
            sleep(1)
            # drecement counter
            n -= 1
        #}
        else: #{
            print("blast off!")
        #}
    #}
    
    def pull_current_time(self): #{
        timestamp = pd.Timestamp.now()
        return timestamp
    #}
    
#}
```
