```python
# IMPORT THE GOODS
import os, sys, time
from time import sleep
from pathlib import Path
import logging
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from random import choice, randint
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
class Logger(): # {
    
    def __init__(self, logging_output_dir): # {
        self.logging_output_dir = logging_output_dir
        # Initiating the Logger object
        self.logger = logging.getLogger(__name__)
        
        # Set the level of the logger (THIS IS SUPER USEFUL since it enables)
        # Explanation regarding the logger levels can be found here:
        # https://docs.python.org/3/howto/logging.html
        self.logger.setLevel(logging.DEBUG)
        
        # Create the logs.log file
        log_file_name = str(pd.Timestamp.now())[:10] + "-ph-viewer-stats"
        log_file_path = os.path.join(logging_output_dir, str(log_file_name) + ".log")
        file_handler = logging.FileHandler(log_file_path)
        
        # Create the console handler with a higher log level
        console_handler = logging.StreamHandler()
        console_handler.setLevel(logging.ERROR)
        
        # Format the logs structure so that every line would include:
        # the time, name level name and log message
        formatter = logging.Formatter('%(asctime)s: %(message)s, FROM=%(funcName)s, LINENO=%(lineno)d, %(name)s - %(levelname)s')
        file_handler.setFormatter(formatter)
        console_handler.setFormatter(formatter)
        
        # Adding the format handlers
        self.logger.addHandler(console_handler)
        self.logger.addHandler(file_handler)
        
        # and printing the logs to the console as well
        self.logger.addHandler(logging.StreamHandler(sys.stdout))
        
    # }
# }
```

```python
class PH_Finale(): # {
    
    def __init__(self, the_logger, function_to_perform): # {
        self.the_logger = the_logger
        self.function_to_perform = str(function_to_perform)
        if self.function_to_perform == "1": # {
            self.the_logger.debug("function 1")
        # }
        elif self.function_to_perform == "2": # {
            self.the_logger.debug("function 2")
        # }
        elif self.function_to_perform == "3": # {
            self.the_logger.debug("function 3")
        # }
        elif self.function_to_perform == "4": # {
            self.the_logger.debug("function 4")
        # }
    # }
    
    def set_up_driver(self): # {
        path_to_chrome_driver = "C:/Chrome/chromedriver.exe"
        return webdriver.Chrome(executable_path=path_to_chrome_driver)
    # }
    
    def tear_down(self, the_driver): # {
        the_driver.quit()
    # }
    
    def countdown(self, n): # {
        # TRY THE FOLLOWING:
        try: # {
            while n >= 0: # {
                sleep(1)
                self.the_logger.debug(str(n))
                n -= 1
            # }
            else: # {
                self.the_logger.info("Blast Off!")
            # }
        # }
        except: # {
            errorMessage = str(sys.exc_info()[0]) + "\n\t\t"
            errorMessage = errorMessage + str(sys.exc_info()[1]) + "\n\t\t"
            errorMessage = errorMessage + str(sys.exc_info()[2]) + "\n"
            exc_type, exc_obj, exc_tb = sys.exc_info()
            fname = os.path.split(exc_tb.tb_frame.f_code.co_filename)[1]
            typeE = str("TYPE : " + str(exc_type))
            fileE = str("FILE : " + str(fname))
            lineE = str("LINE : " + str(exc_tb.tb_lineno))
            messageE = str("MESG : " + "\n" + str(errorMessage) + "\n")
            self.the_logger.error("\n" + typeE +
                          "\n" + fileE +
                          "\n" + lineE +
                          "\n" + messageE)
        # }
        else: # {
            self.the_logger.info("Operation Completed Successfully...")
        # }
    # }
    
    """
    TAKES IN:
    (1) url to HTML webpage
    (2) str of HTML class to search for
    RETURNS:
    list of all contents relative to search
    """
    def pull_html_from_page(self, the_driver, the_html_page, the_class_name): # {
        # USING THE DRIVER, VISIT THE SITE
        the_driver.get(the_html_page)
        # PULL HTML CONTENT
        html_content = the_driver.find_elements(By.CLASS, str(the_class_name))
        return html_content
    # }
    
    """
    TAKES IN:
    (1) url of video link to visit
    (2) file_path of .CSV file to append stats to
    (3?) proxy_to_use ??
    (4?) selenium_driver_to_use ??
    RETURNS:
    none
    """
    def watch_and_log_link(self, the_link_to_watch, the_csv_file): # {
        # TRY THE FOLLOWING
        try: # {
            ############################
            # CREATE TIMESTAMP for START
            time_start = pd.Timestamp.now()
            # append for DataFrame-Entry
            # [2020-01-30]\\ts_list.append(time_start)
            # choose random watch_duration?
            the_watch = randint(75, 180) # WAS: 5, 18
            # SLEEP
            self.countdown(n=the_watch)
            # CREATE timestamp for END
            time_end = pd.Timestamp.now()
            ############################
            # CREATE EMPTY DATAFRAME
            new_entry = pd.DataFrame(data=None, dtype=np.str)
            # CREATE COLUMNS ON "EMPTY" DATAFRAME
            new_entry["Link"] = [str(the_link_to_watch)]
            new_entry["Watch_Duration"] = [str(pd.Timedelta(time_end - time_start))]
            new_entry["Watch_Timestamp"] = [str(time_start)]
            # APPEND TO LOG-FILE/DATAFRAME
            new_entry.to_csv(str(the_csv_file), header=False, index=False, mode='a')
        # }
        except: # {
            errorMessage = str(sys.exc_info()[0]) + "\n\t\t"
            errorMessage = errorMessage + str(sys.exc_info()[1]) + "\n\t\t"
            errorMessage = errorMessage + str(sys.exc_info()[2]) + "\n"
            exc_type, exc_obj, exc_tb = sys.exc_info()
            fname = os.path.split(exc_tb.tb_frame.f_code.co_filename)[1]
            typeE = str("TYPE : " + str(exc_type))
            fileE = str("FILE : " + str(fname))
            lineE = str("LINE : " + str(exc_tb.tb_lineno))
            messageE = str("MESG : " + "\n" + str(errorMessage) + "\n")
            self.the_logger.error("\n" + typeE +
                          "\n" + fileE +
                          "\n" + lineE +
                          "\n" + messageE)
        # }
        else: # {
            self.the_logger.info("Operation Completed Successfully...")
        # }
    # }
# }
```

```python
if __name__ == "__main__": # {
    # SETUP LOGGER
    # [2020-01-30]\\logger = Logger(logging_output_dir="C:/data/outbound/").logger
    logger = Logger(logging_output_dir=".").logger
    # TEST LOGGINGS
    # [2020-01-30]\\logger.debug("log DEBUG")
    # [2020-01-30]\\logger.error("log ERROR")
    ###############
    test_1 = PH_Finale(the_logger=logger, function_to_perform=1)
    test_list = ["https://www.reddit.com/r/gwcouples", 
                 "https://www.google.com/", 
                 "https://www.bing.com/"]
    # LOOP THRU LIST
    for item in test_list: # {
        test_1.watch_and_log_link(the_link_to_watch=str(item),
                                  the_csv_file="ph_urls_visited.csv")
    # }
    
# }
```
