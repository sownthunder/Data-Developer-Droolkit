```python
import pyautogui
pyautogui.FAILSAFE = True
pyautogui.PAUSE = 0.5
```

```python
from pynput import mouse
```

```python
def on_move(x, y): #{
    print('Pointer moved to {0}'.format(
        (x, y)))
#}
```

```python
def on_click(x, y, button, pressed): #{
    # RE-INSTANTIATE GLOBALS
    global initial_x_pos, initial_y_pos
    print('{0} at {1}'.format(
        'Pressed' if pressed else 'Released',
        (x, y)))
    # assign to GLOBALS
    initial_x_pos = int(x)
    initial_y_pos = int(y)

    if not pressed: #{
        # Stop listener
        return False
    #}
#}
```

```python
def on_scroll(x, y, dx, dy): #{
    print('Scrolled {0} at {1}'.format(
        'down' if dy < 0 else 'up',
        (x, y)))
#}
```

```python
# Collect events until released
with mouse.Listener(
    on_move=on_move,
    on_click=on_click,
    on_scroll=on_scroll) as listener:
    listener.join()



sleep(15)
pyautogui.moveTo(x=initial_x_pos, y=initial_y_pos, duration=5.0)
```
