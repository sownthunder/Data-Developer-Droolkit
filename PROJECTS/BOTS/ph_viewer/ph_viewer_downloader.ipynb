## Setup LOGGER

```python
def setup_logger(): #{
    # RE-INSTANTIATE GLOBAL
    global timestamp_str
    # TRY THE FOLLOWING
    try: #{
        logging.basicConfig(level=logging.INFO,
                           format="%(asctime)s:%(message)s",
                           filename="C:/data/Logs/PH-VIEWER-downloader" + timestamp_str + ".log",
                           filemode="a")
    #}
    except: #{
        print("failed setting up logger yo")
    #}
    else: #{
        print("SUCCESS! VERY NICE!")
    #}
#}
```

## MAKE DIRECTORY FOR TODAYS DATE:

```python
if __name__ == "__main__": #{
    # INSTANTIATE GLOBAL VARIABLES
    out_directory = "C:/data/outbound/ph_viewer"
    ts_date_str = str(pd.Timestamp.now())[:10]
    ts_time_str = str(pd.Timestamp.now())[11:19]
    ts_time_str = ts_time_str.replace(":","-")
    timestamp_str = ts_date_str + "_" + ts_time_str
    print(ts_date_str)
    print(ts_time_str)
    print(timestamp_str)
    ###########################################################
    setup_logger()
    #######################################################
    # CHECK AND CREATE NEW DIR PATH
    new_dir_path = os.path.join(out_directory, str(ts_date_str + "-" + str(ts_time_str)))
    # IF 'NEW DIR' DOES NOT EXIST (it wont)
    if not os.path.exists(new_dir_path): #{
        # MAKE IT EXIST!
        os.makedirs(new_dir_path)
    #}
    
    options = webdriver.ChromeOptions()
    options.add_argument('--ignore-certificate-errors')
    options.add_argument("--test-type")
    #options.binary_location = "/usr/bin/chromium/"
    driver = webdriver.Chrome(chrome_options=options)
    
    driver.get('https://pornhub.com/')
    
    images = driver.find_elements_by_tag_name('img')
    
    # COUNTER
    i = 1
    for image in images: #{
        logging.info(image.get_attribute('src'))
        print(str(image.get_attribute('src')))
        # GET IMAGE SOURCE
        src = image.get_attribute('src')
        # CHECK STRING OF "src"
        if str(src) != "None": #{
             # CREATE FILENAME
            file_name = ts_date_str + "-" + ts_time_str + str(i) + ".png"
            # CREATE FILE_PATH
            file_path = os.path.join(new_dir_path, file_name)
            print("FILEPATH WILL BE ===" + str(file_path))
            # TRY THE FOLLOWING
            try: #{
                # DOWNLOAD IMAGE
                urllib2.urlretrieve(str(src), file_path)
            #}
            except: #{
                # TRY DOWNLOADING VIA WGET
                wget.download(src, file_path)
            #}
            else: #{
                print("SUCCESSFULLY DOWNLOADED VIA urllib!")
            #}
            finally: #{
                pass    
            #}
        #}
        else: #{
            print("FAIL!")
        #}
        # INCREASE COUNT
        i += 1
        logging.info(str(i))
        print(str(i))
    #}
    # QUIT OUT OF DRIVER
    driver.close()
    # END THE PROGRAM
    sys.exit(69)
#}
```
