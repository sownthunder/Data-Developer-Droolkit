```python
import os, sys, time
from time import sleep
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
```

```python
def test_command(): #{
    print("TESTED AT: \t" + str(pd.Timestamp.now())[:10])
#}
```

```python
if __name__ == "__main__": #{
    # MAIN APP PROPERTIES 
    root = tk.Tk()
    root.title("CofA-Crate")
    root.geometry("550x205+250+250")
    root.minsize(width=450, height=105)
    root.maxsize(width=750, height=405)
    
    #######################################################
    # TOP FRAME
    topframe = tk.Frame(master=root)
    topframe.pack(anchor=tk.N, fill=tk.BOTH, expand=True)
    
    ######################
    # b1
    button1 = tk.Button(master=topframe, text="b1", background="#ffff66",
                       activebackground="#000000", activeforeground="#666666",
                        disabledforeground="#012562", command=test_command)
    button1.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    
    #######################
    # e1
    entry1 = tk.Entry(master=topframe, text="entry1", font=("Sourcecode Semibold", 14))
    entry1.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)
    
    #######################################################
    # MIDDLE FRAME
    middleframe = tk.Frame(master=root)
    middleframe.pack(anchor=tk.CENTER, fill=tk.BOTH, expand=True)
    
    #####################
    # b2
    button2 = tk.Button(master=middleframe, text="b2", background="#ffff66",
                       activebackground="#000000", activeforeground="#666666",
                        disabledforeground="#012562", command=test_command)
    button2.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    
    ######################
    # e2
    entry2 = tk.Entry(master=middleframe, text="entry2", font=("Sourcecode Semibold", 14))
    entry2.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)
    
    #######################################################
    
    # BOTTOM FRAME
    bottomframe = tk.Frame(master=root)
    bottomframe.pack(anchor=tk.S, fill=tk.BOTH, expand=True)
    
    ########################################################
    
    # BOTTOM_LEFT
    bottomleft = tk.Frame(master=bottomframe)
    bottomleft.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    
    # b3
    button3 = tk.Button(master=bottomleft, text="Confirm", background="#ffff66",
                       activebackground="#000000", activeforeground="#666666",
                        disabledforeground="#012562")
    button3.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    
    # b4
    button4 = tk.Button(master=bottomleft, text="Cancel", background="#ffff66",
                       activebackground="#000000", activeforeground="#666666",
                        disabledforeground="#012562")
    button4.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    
    ########################################################
    
    # BOTTOM_RIGHT
    bottomright = tk.Frame(master=bottomframe)
    bottomright.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)
    
    # R1
    radio1 = tk.Radiobutton(master=bottomright, text="Current **ONLY** CofAs", background="#ffff66",
                           activebackground="#000000", activeforeground="#666666",
                           disabledforeground="#012562")
    radio1.pack(side=tk.TOP, fill=tk.BOTH, expand=True)
    
    # R@
    radio2 = tk.Radiobutton(master=bottomright, text="Return **ALL** CofAs", background="#fff666",
                           activebackground="#ffff66", activeforeground="#ff6666",
                           disabledforeground="#012562")
    radio2.pack(side=tk.BOTTOM, fill=tk.BOTH, expand=True)
    
    root.config()
    root.mainloop()
#}
```
