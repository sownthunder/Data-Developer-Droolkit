```python
def setup_logger(): #{
    logging.basicConfig(level=logging.INFO,
                       format='%(asctime)s: %(message)s',
                       filemode='a')
#}
```

```python
def scan_directory(is_single_dir, dir_to_scan): #{
    # RE-INSTANTIATE GLOBALS
    global out_directory, outbound_directory
    # TRY THE FOLLOWING
    try: #{
        logging.info("BEGIN SCANNING SCRIPT >>> " + str(dir_to_scan) + "\n")
        logging.info("outboundDirectory == " + str(outbound_directory) +"\n")
        logging.info("inDirectory == " + str(dir_to_scan) + "\n")
        is_directory = dir_to_scan.is_dir()
        logging.info("is a Dir == " + str(is_directory) + "\n")
        # CREATE LIST OF STRINGS
        doc_list = os.listdir(dir_to_scan)
        logging.info("length of [doc_list]: \t" + str(len(doc_list)) + "\n")
        # CREATE SERIES OFF LIST
        d1 = pd.Series(doc_list)
        d1.astype(dtype=np.str)
        # CREATE EMPTY LISTS TO STORE DATA
        pn_list = []
        lang_list = []
        country_list = []
        file_list = []
        date_list = []
        logging.info("CREATING [dataframe]...")
        data_frame = pd.DataFrame(data=None, dtype=np.str)
        # COUNTER-BABY
        x = 0
        # ITERATE THRU DIRECTORY 
        for root, d_names, f_names in os.walk(dir_to_scan): #{
            logging.info("<BEGIN_LOOP # " + str(x) + "> ==================<> \n|")
            # IF THERE ARE FILES IN DIRECTORY
            if len(f_names) >= 0: #{
                # FOR EACH FILE <<PERFORM THE FOLLOWING ACTIONS>>
                for f in f_names: #{
                    the_str = str(f)
                    the_len = len(the_str)
                    logging.info("TEST-1== " + str(root) + "\\" + "_PART_NO_.PDF")
                    # SUB-STR OPERATIONS FOR SDS-PN
                    logging.info("TEST-2== " + str(root) + "\\" + the_str + "\n")
                    index_mrk = the_str.find('_', 0, the_len)
                    logging.info("| ROOT == " + str(root))
                    pn_str = str(the_str[0:index_mrk])
                    logging.info("| SDS-PN == " + pn_str)
                    pn_list.append(pn_str)
                    ##########################################
                    # SUB STR OPERATIONS FOR LANGUAGE & LANG #
                    # === GET COUNTRY BY USING OS.PATH.XXX   #
                    ##########################################
                    country_index_mrk = root.rfind('\\', 0, len(root))
                    logging.info("* index_mark == " + str(country_index_mrk))
                    country_str = str(root[country_index_mrk + 1:len(root)])
                    logging.info("| COUNTRY == " + country_str)
                    country_list.append(country_str)
                #}
            #}
        #}
    #}
    except: #{
        errorMessage = str(sys.exc_info()[0]) + "\n\t\t"
        errorMessage = errorMessage + str(sys.exc_info()[1]) + "\n\t\t"
        errorMessage = errorMessage + str(sys.exc_info()[2]) + "\n"
        exc_type, exc_obj, exc_tb = sys.exc_info()
        fname = os.path.split(exc_tb.tb_frame.f_code.co_filename)[1]
        typeE = str("TYPE : " + str(exc_type))
        fileE = str("FILE : " + str(fname))
        lineE = str("LINE : " + str(exc_tb.tb_lineno))
        messageE = str("MESG : " + str(errorMessage) + "\n")
        logging.error("\n" + typeE +
              "\n" + fileE +
              "\n" + lineE +
              "\n" + messageE)
    #}
    else: #{
        logging.info("SUCCESS! VERY NICE!\n")
    #}
#}
```

```python
if __name__ == "__main__": #{
    setup_logger()
    # INSTANTIATE GLOBAL VARIABLES
    in_file = Path("C:/data/inbound/ARI_ChemGes SDS - Available Languages&Countries.csv")
    file_conventions = pd.read_csv(in_file, index_col='File Naming Convention', )
    print(file_conventions)
    in_directory = Path("J:/controlled_docs/SDS/Agilent_SDS/")
    in_directory2 = Path("J:/controlled_docs/SDS/ALL SDS's/")
    out_directory = Path("C:/Temp/")
    outbound_directory = Path("J:/controlled_docs/SDS/SDS Lists/")
    scan_directory(is_single_dir=False, dir_to_scan=in_directory)
    scan_directory(is_single_dir=True, dir_to_scan=in_directory2)
#}
```
