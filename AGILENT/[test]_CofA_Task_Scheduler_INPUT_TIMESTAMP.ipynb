## *DEFINE FUNCTIONS*

```python
def countdown(n): #{
    while n > 0 : #{
        logging.info(n)
        sleep(1) # sleep for 1 second
        n -= 1
    #}
    else: #{
        logging.info("BLAST OFF!")
    #}
#}
```
