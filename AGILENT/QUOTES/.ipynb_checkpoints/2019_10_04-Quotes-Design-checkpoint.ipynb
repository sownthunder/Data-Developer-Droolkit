```python
import os, sys, time
from time import sleep
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
```

```python
if __name__ == "__main__": #{
    # MAIN APP PROPERTIES
    root = tk.Tk()
    root.title("Quotes")
    root.geometry("850x505+250+250")
    root.minsize(width=850, height=505)
    root.maxsize(width=1050, height=705)
    
    onevar = tk.BooleanVar()
    twovar = tk.BooleanVar()
    threevar = tk.BooleanVar()
    
    #############
    # TOP FRAME #
    #############
    topframe = tk.Frame(master=root)
    topframe.pack(anchor=tk.N, fill=tk.BOTH, expand=True)
    
    ######
    # e1 #
    e1 = tk.Entry(master=topframe, font=("Courier New", 18))
    e1.pack(side=tk.LEFT, fill=tk.X, expand=True)
    ################################################
    
    ######
    # e2 #
    e2 = tk.Entry(master=topframe, font=("Courier New", 18))
    e2.pack(side=tk.RIGHT, fill=tk.X, expand=False)
    ################################################
    
    ################
    # MIDDLE FRAME #
    ################
    middleframe = tk.Frame(master=root)
    middleframe.pack(anchor=tk.CENTER, fill=tk.BOTH, expand=True)
    
    ###############
    # MIDDLE-LEFT #
    ###############
    middleleft = tk.Frame(master=middleframe)
    middleleft.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    
    ############
    # TREEVIEW #
    tree=ttk.Treeview(master=middleleft)
    tree["columns"]=("one","two","three")
    # DEFINITION OF THE COLUMNS
    tree.column("one", width=250, minwidth=200)
    # [2019-10-04]... tree.column("two", width=400, minwidth=200)
    tree.column("two", width=150)
    tree.column("three", width=150)
    # DEFINITION OF THE HEADINGS
    tree.heading("one", text="Subject")
    tree.heading("two", text="From")
    tree.heading("three", text="Recieved")
    # INSERT SOME ROWS
    tree.insert("" , 0,    text="Email Box:", values=("1A","1b"))
    id2 = tree.insert("", 1, "dir2", text="Inbox") # Dir 2
    tree.insert(id2, "end", "dir 2", text="sub dir 2", values=("2A","2B"))
    
    ##alternatively:
    tree.insert("", 3, "dir3", text="Archive") # Dir 3
    tree.insert("dir3", 3, text=" sub dir 3",values=("3A"," 3B"))
    # PACKING
    tree.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    #####################################################
    
    ################
    # MIDDLE-RIGHT #
    ################
    middleright = tk.Frame(master=middleframe)
    middleright.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    
    ######
    # c1 #
    c1 = ttk.Checkbutton(master=middleright, text="one",
                         variable=onevar, onvalue=True)
    c1.pack(anchor=tk.N, fill=tk.Y, expand=False)
    
    ######
    # c2 #
    c2 = ttk.Checkbutton(master=middleright, text="two",
                        variable=twovar, onvalue=True)
    c2.pack(anchor=tk.CENTER, fill=tk.Y, expand=True)
    
    ######
    # c3 #
    c3 = ttk.Checkbutton(master=middleright, text="three",
                        variable=threevar, onvalue=False)
    c3.pack(anchor=tk.S, fill=tk.Y, expand=True)
    
    ###############
    # BOTTOMFRAME #
    ###############
    bottomframe = tk.Frame(master=root)
    bottomframe.pack(anchor=tk.S, fill=tk.BOTH, expand=True)
    
    ######
    # b1 #
    b1 = tk.Button(master=bottomframe, text="SUBMIT", 
                   font=("Wingdings", 20), relief=tk.GROOVE)
    b1.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    
    ######
    # b2 #
    b2 = tk.Button(master=bottomframe, text="CANCEL", 
                   font=("Wingdings", 20), relief=tk.RIDGE)
    b2.pack(side=tk.RIGHT, fill=tk.BOTH, expand=False)
    
    root.config()
    root.mainloop()
    
#}
```
