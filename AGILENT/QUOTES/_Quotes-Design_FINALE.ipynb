```python
def populate_tree(the_tree): #{
    # RE-INSTANTIATE GLOBALS
    global root
    # VIEW RECORDS #####
    items = the_tree.get_children()
    for item in items: #{
        the_tree.delete(item)
    #}
    ###################################################
    query = "SELECT * FROM quotes ORDER BY [Timestamp]" # # name desc
    ###################################################
    # TRY THE FOLLOWING
    try: #{
        # CREATE THE CONNECTION
        conn = sqlite3.connect('test_quotes.db')
        print("Opened database successfully")
        # [2019-10-31]\\cursor = conn.cursor()
        # [2019-10-31]\\query_result = cursor.execute(query, "")
        cursor = conn.execute(query)
        for row in cursor: #{
            print("Tracking#  = " + str(row[0]))
            print("Name = " + str(row[1]))
            print("Email = " + str(row[2]))
            print("Type = " + str(row[3]))
            print("Timestamp = " + str(row[4]))
            print("Sent = " + str(row[5]))
            # CREATE LIST
            test_lst = [str(row[1]), str(row[2]), str(row[3]), str(row[4]), str(row[5])]
            # INSERT INTO TREE
            the_tree.insert('', 0, text=str(row[0]), values=test_lst)
        #}
        print("Operation done successfully")
        conn.close()
    #}
    except: #{
        print("FAILURE IN THE TREE-VIEW!")
    #}
    else: #{
        print("SUCCESS! VERY NICE!")
    #}
#}
```
