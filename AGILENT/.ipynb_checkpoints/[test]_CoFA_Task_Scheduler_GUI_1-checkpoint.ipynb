```python
import os, sys, time
from time import sleep
from pathlib import Path
import logging, subprocess
import tkinter as tk
from tkinter import ttk
from tkinter import dnd
from tkinter import filedialog
from tkinter import simpledialog, commondialog
from tkinter import colorchooser
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
def App(): #{
    # RE-INSTANTIATE GLOBALS
    global root, entry_var, hour_var
    root.title("CofA Task Scheduler")
    root.geometry("520x105+300+300")
    root.resizable(width=False, height=True)
    root.executable_1 = filedialog.askopenfilename(
        initialdir = "/",
        title = "Select file",
        filetypes = (("exe files","*.exe"),
                     ("all files","*.*")))
    print(str(root.executable_1))
    entry_var.set(str(root.executable_1))
    
    # TOP FRAME
    topframe = tk.Frame(master=root)
    topframe.pack(anchor=tk.N, fill=tk.BOTH, expand=True)
    
    l1 = tk.Label(master=topframe, font=("Courier New", 12), text="Command-1:")
    l1.pack(anchor=tk.CENTER, fill=tk.X, expand=True)
    
    e1 = tk.Entry(master=topframe, textvariable=entry_var)
    e1.pack(anchor=tk.CENTER, fill=tk.X, expand=True)
    
    # BOTTOM FRAME
    bottomframe = tk.Frame(master=root)
    bottomframe.pack(anchor=tk.S, fill=tk.BOTH, expand=True)
    
    l2 = tk.Label(master=bottomframe, font=("Courier New", 12), text="Hour-Set (24-hour):")
    l2.pack(anchor=tk.CENTER, fill=tk.X, expand=True)
    
    w = tk.Spinbox(master=bottomframe, from_ = 0, to = 24, textvariable=hour_var)
    w.pack(anchor=tk.CENTER, fill=tk.X, expand=True)
    
    b1 = tk.Button(master=bottomframe, font=("Courier New", 12), text="run", relief=tk.GROOVE)
    b1.pack(anchor=tk.CENTER, fill=tk.X, expand=True)
    
    root.config()
    root.mainloop()
#}
```

```python
def countdown(n): #{
    while n > 0 : #{
        print(n)
        sleep(1) # sleep for 1 second
        n -= 1
        if n == 0: #{
            print("BLAST OFF!")
        #}
    #}
#}
```

```python
def main(): #{
    pass
#}
```

```python
if __name__ == "__main__": #{
    # INSTANTIATE GLOBAL VARIABLES
    root = tk.Tk()
    entry_var = tk.StringVar()
    hour_var = tk.IntVar()
    hour_var.set(int(12))
    ##############################
    App()
#}
```
