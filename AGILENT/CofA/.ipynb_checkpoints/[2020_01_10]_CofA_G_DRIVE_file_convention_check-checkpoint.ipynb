this attempts to fix the fact that `G:/C of A's/Agilent/*` does not all fall under the same file convention:

```python
import os, sys, time
from time import sleep
from pathlib import Path
from fnmatch import fnmatch as mcmatchy
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import logging
```

```python
"""
TAKES IN path to pdf and edits the STRING
of that PDF to match the naming convention
in G:/C of A's/#Email Node/ from F:/APPS/CofA/
"""
def generate_naming_convention(the_pdf_path): #{
    # get/set filename to variable
    the_file_name = str(os.path.basename(the_pdf_path))
    # PERFORM STRING OPERATIONS
    #################################
    idx_mrk = the_file_name.rfind('@', 0, len(the_file_name))
    half1 = str(the_file_name[0:idx_mrk])
    half2 = str(the_file_name[idx_mrk + 1:len(the_file_name)])
    logging.debug("HALF 1 == " + half1)
    logging.debug("HALF 2 == " + half2)
    #  setup NEW FILE NAME (for copy)
    new_name = "part "
    new_name += str(half1)
    new_name += " CofA Lot# "
    new_name += str(half2)
    logging.debug("NEW NAME == " + str(new_name))
    #################################
    return str(new_name)
#}
```

```python
def main():  # {
    # COUNTER (total # of possible files)
    x = 0
    # RE-INSTANTIATE GLOBALS
    global in_directory, ignore_list_f, ignore_list_g, outlier_list, proper_fn_list
    # BEGIN FUNCTIONS
    for root, dirs, files in os.walk(in_directory):  # {
        # FOR EACH ITEM IN IGNORE LIST...
        for item in ignore_list_g:  # {
            if str(item) in dirs:  # {
                # REMOVE FROM OS.WALK
                dirs.remove(str(item))
            #}
        #}
        # FOR EACH FILE...
        for f in files:  # {
            # MAKE SURE WE ARE EVEN LOOKING AT A PDF
            if mcmatchy(f, "*.pdf"): # {
                # CHECK AND SEE IF IT FALLS UNDER CONVENTION...
                if mcmatchy(f, "part*.pdf") or mcmatchy(f, "*Archive*"):  # {
                    print("NAMING CONVENTION!\t\t\t<> CHECK!")
                # }
                elif mcmatchy(f, "*CofA Lot*"): # {
                    # STILL AN OUTLIER
                    file_path = os.path.join(root, f)
                    print("\n================================================\n" 
                          + str(file_path) + "\t\t\t<> FAIL !\n")
                    # ADD OUTLIER TO LIST
                    outlier_list.append(str(file_path))
                    # CREATE FILE NAME CONVENTION STR
                    file_name_conv = generate_naming_convention(the_pdf_path=f)
                    print("\t\t [FILE_NAME_CONV] == " + str(file_name_conv))
                    # CREATE PROPER 'old_path' OFF OF FILE_NAME_CONV
                    test_old_path = os.path.join(root, file_name_conv)
                    print("TEST: \t " + str(test_old_path))
                    # ADD PROPER NAMING TO LIST
                    proper_fn_list.append(str(file_name_conv))
                #}
                else:  # {
                    file_path = os.path.join(root, f)
                    print("\n================================================\n" 
                          + str(file_path) + "\t\t\t<> FAIL !\n")
                    # ADD OUTLIER TO LIST
                    outlier_list.append(str(file_path))
                    # CREATE FILE NAME CONVENTION STR
                    file_name_conv = generate_naming_convention(the_pdf_path=f)
                    print("\t\t [FILE_NAME_CONV] == " + str(file_name_conv))
                    # CREATE PROPER 'old_path' OFF OF FILE_NAME_CONV
                    test_old_path = os.path.join(root, file_name_conv)
                    print("TEST: \t " + str(test_old_path))
                    # ADD PROPER NAMING TO LIST
                    proper_fn_list.append(str(file_name_conv))
                # }
            # }
            else: # {
                print("NOT A PDF!\t\t\t\t<> FAIL !")
            # }
            # increment counter
            x += 1 
        # }
    # }
    print("LENGTH == " + str(len(outlier_list)) + " outliers...")
    print("OUT OF == " + str(x) + "total...")
# }
```

```python
if __name__ == "__main__": # {
    # INSTANTIATE GLOBAL VARIABLES
    in_directory = Path("G:/C of A's/Agilent/")
    # DIRECTORIES TO IGNORE (F:/APPS/CofA/)
    ignore_list_f = ['Archive ERR',
                   'Archive - For all archived CofA, see G CofA folder',
                   'Instruction Sheets',
                   'EXPORT ERRORS']
    # DIRECTORIES TO IGNORE (G:/C of A's/Agilent/)
    ignore_list_g = ['#Archive']
    # EMPTY LIST TO HOLD OUR OUTLIERS
    outlier_list = []
    # EMPTY LIST TO HOLD **PROPER** FILE NAME CONVENTIONS
    proper_fn_list = []
    # CALL MAIN FUNCTION
    main()
#}
```

```python
export_df = pd.DataFrame(data=None, dtype=np.str)
# ADD IN COLUMNS TO DATAFRAME
export_df["Outlier"] = outlier_list
export_df["proper_name"] = proper_fn_list
file_name_str = str("C:/data/outbound/2019-12-24/" 
                    + "outliers-" 
                    + str(pd.Timestamp.now())[:10] 
                    + ".csv")
export_df.to_csv(Path(file_name_str), index=False)
```
