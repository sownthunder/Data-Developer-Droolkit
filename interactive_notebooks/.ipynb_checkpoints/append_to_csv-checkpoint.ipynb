```python
import os, sys, time
from time import sleep
from pathlib import Path
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
def countdown(n):  # {
    while n >= 0:  # {
        print(n)
        sleep(1)
        n -= 1
    # }
    else: # { 
        print("BLAST OFF!")
    # }
# }
```

```python
def fake_video_watch(sleep_time, number_of_loops):  # {
    # RE-INSTANTIATE GLOBALS
    global in_file
    print("sleep_time:\t" + str(sleep_time))
    print("num_of_loops:\t" + str(number_of_loops) + "\n=====================\n")
    # COUNTER
    i = 0
    while i < number_of_loops:  # {
        print("loops remaining: " + str(int(number_of_loops) - i))
        test_link = ["https://www.google.com"]
        test_count = [29]
        test_views = ["29"]
        # [2019-12-23]\\test_time = [str(pd.Timestamp.now())[:16]]
        test_time = [str(pd.Timestamp.now())]
        # CREATE DATAFRAME
        df_test = pd.DataFrame(data=None, dtype=np.str)
        # APPEND NEW COLUMN ENTRIES TO DATAFRAME
        df_test["link"] = test_link
        df_test["count"] = test_count
        df_test["views"] = test_views
        df_test["time"] = test_time
        countdown(int(sleep_time))
        print("\n\t\t writing to file...")
        # append to file because we have watched!
        df_test.to_csv(in_file, header=None, index=False, mode='a')
        # increment counter 
        i += 1
    #}
#}
```

```python
if __name__ == "__main__": # {
    # TRY THE FOLLOWING
    try:  # {
        # INSTANTIATE GLOBAL VARIABLES
        in_file = Path("append_test.csv")
        # call function
        fake_video_watch(sleep_time=10, number_of_loops=5)
    #}
    except: #{
        pass
    #}
    else:  # {
        print("Operation Completed Successfully...")
    #}
# } 
```
